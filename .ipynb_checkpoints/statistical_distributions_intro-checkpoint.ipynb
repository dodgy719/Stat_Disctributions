# Important concept of Numpy random seed

NumPy random seed is simply a function that sets the random seed of the NumPy pseudo-random number generator. It provides an essential input that enables NumPy to generate pseudo-random numbers for random processes.

**Now what is a pseudo-random number?**

Pseudo-random numbers are computer generated numbers that appear random, but are actually predetermined.

**Why are they needed?**

if you give a computer a certain input, it will precisely follow instructions to produce an output. And if you later give a computer the same input, it will produce the same output.

Importantly, because pseudo-random number generators are deterministic, they are also repeatable.
If you run the algorithm with the same input, it will produce the same output.

## Probability Mass Functions (PMFs)

Probability Mass Function: think of it like masses hanging on a bar

PMF <--> Histogram

### Visual

alt text

### Coding Example


```python
import matplotlib.pyplot as plt
import collections
import numpy as np
%matplotlib inline
```

```python
roll_die1 = np.random.randint(1, 7, size=(3000)) 
roll_die2 = np.random.randint(1, 7, size=(3000))
sum_of_dice = roll_die1 + roll_die2

print(roll_die1)
print(roll_die2)
print(sum_of_dice)
```

```python
counter_die1 = collections.Counter(roll_die1)
counter_die2 = collections.Counter(roll_die2)
counter_sum = collections.Counter(sum_of_dice)
print(f'The list of events: {sum_of_dice}')
print(counter_sum)

```

```python

plot_range = range(1,13)

pmf_die1 = [round(counter_die1.get(k,0)/len(roll_die1), 2) for k in plot_range]
pmf_die2 = [round(counter_die2.get(k,0)/len(roll_die2), 2) for k in plot_range]
pmf_sum = [round(counter_sum.get(k,0)/len(sum_of_dice), 2) for k in plot_range]
```

```python
# Plotting parts
plt.style.use('fivethirtyeight')

plt.title ("Probability Mass Function (Die #1)")
plt.stem(plot_range, pmf_die1)
plt.show()

plt.title ("Probability Mass Function (Die #2)")
plt.stem(plot_range, pmf_die2)
plt.show()
```

```python
plt.title ("Probability Mass Function For Dice Summed")
plt.stem(plot_range, pmf_sum)
plt.show()

plt.hist(sum_of_dice, range=(1,12))
plt.show()
```

## Cummulative Distribution Functions (CDFs)

Easier way to show many possibilites (compared to PMF)

$$ F(x) = P(X\leq x) $$

```python
plt.step(range(1,13), np.cumsum(pmf_die1))
```
